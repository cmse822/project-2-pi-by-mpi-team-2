```python
import seaborn as sns
sns.set_theme(style="darkgrid")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read DataFrame 
df = pd.read_csv("data/Part4_Results.csv",names=["num_darts","size_round","pi_result","total_runtime","num_processors"])
df
```

```python
# Calculate Error
df["error"] = np.abs(df["pi_result"] - np.pi)

# Plot results 
fig, ax = plt.subplots()

for processors in np.unique(df["num_processors"]):
    idx = (df["num_processors"] == processors) & (df["size_round"] == 100)

    plt.plot(df["num_darts"].loc[idx], df["error"].loc[idx])
plt.yscale('log')
plt.xscale('log')
plt.xlabel("Number of darts")
plt.ylabel("Error")
plt.legend(np.unique(df["num_processors"]).astype("str"),title="Number of processor")
```
